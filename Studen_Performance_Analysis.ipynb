```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student.csv')
df.head()
```

```python
df['Gender'] = df['Gender'].replace({0: 'Male', 1: 'Female'}) # Replacing gender values
df['Extracurricular'] = df['Extracurricular'].replace({0: 'NO', 1: 'YES'}) # Replacing extracurricular values
```

```python
# Rounding off all numeric columns to 2 decimal places
df = df.round(2)
df.head()
```

```python
# Box plot for 'Gender' vs 'GPA'
sns.boxplot(x='Gender', y='GPA', data=df)
plt.xlabel('Gender')
plt.ylabel('GPA')
plt.title('Gender vs GPA')
plt.show()
```

```python
# Bar plot for 'Extracurricular' vs 'GPA'
sns.barplot(x='Extracurricular', y='GPA', data=df)
plt.xlabel('Extracurricular Activities')
plt.ylabel('GPA')
plt.title('Extracurricular Activities vs GPA')
plt.show()
```

```python
# Histogram for 'Age'
plt.hist(df['Age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
# Scatter plot for 'Study Hours' vs 'GPA'
plt.scatter(df['StudyTimeWeekly'], df['GPA'])
plt.xlabel('Study Hours')
plt.ylabel('GPA')
plt.title('Study Hours vs GPA')
plt.show()
```
